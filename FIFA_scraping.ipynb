```python
def get_matches(year):
  site = f"https://en.wikipedia.org/wiki/{year}_FIFA_World_Cup"
  response = requests.get(site)
  contenido = response.text
  #lxml es el parser a utilizar para leer html
  soup = BeautifulSoup(contenido, "lxml")
  partidos = soup.find_all("div", class_ = "footballbox")
  #armamos 3 listas, equipo local, score y visita
  home = []
  score = []
  away= []
  for partido in partidos:
    home.append(partido.find("th", class_= "fhome").get_text())
    score.append(partido.find("th", class_= "fscore").get_text())
    away.append(partido.find("th", class_= "faway").get_text())
  dict_football = {"local": home, "score": score, "visita": away}
  df_football = pd.DataFrame(dict_football)
  df_football["year"] = year
  return df_football
#para indicar el año que quiero correr
print(get_matches(1986))
print(df_football)
```

```python
#para scrapear los resultados de TODOS los mundiales genero una función
years = [1930, 1934, 1938, 1942, 1946, 1950, 1954, 1958, 1962]
#creo una lista con TODOS los df de c/u de los mundiales
fifa = [get_matches(year) for year in years]
 #concateno todos los df listados en fifa... del punto anterior
df_fifa = pd.concat(fifa, ignore_index = True)
df_fifa.to_csv("fifa_world_cup.csv", index=False)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
#soluciona problemas con max excep SSL
from urllib3.exceptions import InsecureRequestWarning
from urllib3 import disable_warnings
disable_warnings(InsecureRequestWarning)
years = [1930, 1934, 1938, 1950, 1954, 1958, 1962, 1966, 1970, 1974, 1978,
1982, 1986, 1990, 1994, 1998, 2002, 2006, 2010, 2014, 2018]
def obtener_partidos(year):
  web = f"https://en.wikipedia.org/wiki/{year}_FIFA_World_Cup"
  response = requests.get(web, verify=False)
  contenido = response.text
  soup = BeautifulSoup(contenido, "lxml") #además hay q instalar el lxml con pip install lxml
  partidos = soup.find_all("div", class_ = "footballbox")
  home = []
  away = []
  score = []
  for partido in partidos:
    home.append(partido.find("th", class_ = "fhome").get_text())
    away.append(partido.find("th", class_ = "faway").get_text())
    score.append(partido.find("th", class_ = "fscore").get_text())
  diccionario_football = {"home": home, "score": score, "away":away }
  df_football = pd.DataFrame(diccionario_football)
  df_football["year"] = year
  return df_football
#genero una lista en base a un bucle for para colectar la data histórica
fifa = [obtener_partidos(year) for year in years]
#genero el df concatenando todos los registros de la lista
df_fifa = pd.concat(fifa, ignore_index=True)
#genero un csv con el df llamado fifa_world_cup
#ruta_csv =  "\historico_fifa_world_cup.csv"
# guardar DataFrame en archivo CSV
#df_fifa.to_csv(path_or_buf=ruta_csv, index=False
df_fifa
```
